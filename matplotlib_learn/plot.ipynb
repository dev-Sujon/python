# What is Matplotlib?

Matplotlib is a low level graph plotting library in python that serves as a visualization utility.

Matplotlib was created by John D. Hunter.

Matplotlib is open source and we can use it freely.

Matplotlib is mostly written in python, a few segments are written in C, Objective-C and Javascript for Platform compatibility.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(8, 6))

plt.plot(x, y, label='Solid Line', linestyle='-')
plt.plot(x, y + 0.5, label='Dashed Line', linestyle='--')
plt.plot(x, y + 1.0, label='Dotted Line', linestyle=':')
plt.plot(x, y + 1.5, label='Dash-Dot Line', linestyle='-.')
plt.plot(x, y - 0.5, label='No Line', linestyle='')

plt.legend()
plt.title('Matplotlib Linestyle Types')
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(y,x,linewidth=2.0,linestyle="--",marker='d')
```

```python
import matplotlib.pyplot as plt

# Create a figure and axis
fig, ax = plt.subplots()

# Example 1: Plot a point (circle)
ax.plot(0, 0, 'o', markersize=10, label='Point (circle)')

# Example 2: Plot a line (rectangle)
ax.plot([1, 2, 2, 1, 1], [1, 1, 2, 2, 1], label='Line (rectangle)')

# Example 3: Plot a polygon (triangle)
ax.plot([3, 4, 5, 3], [1, 2, 1, 1], label='Polygon (triangle)')

# Example 4: Plot a circle
circle = plt.Circle((7, 1.5), 1, fill=False, label='Circle')
ax.add_patch(circle)

# Example 5: Plot an ellipse
ellipse = plt.Ellipse((9, 1.5), 2, 1, fill=False, label='Ellipse')
ax.add_patch(ellipse)

# Set axis limits and labels
ax.set_xlim(0, 11)
ax.set_ylim(0, 3)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

# Add legend
ax.legend()

# Show the plot
plt.show()
```

# circle

```python
circle_radius = 0.5
circle_theta = np.linspace(0, 2 * np.pi, 100)
circle_x = circle_radius * np.cos(circle_theta) + 3
circle_y = circle_radius * np.sin(circle_theta) + 1.5

plt.plot(circle_x, circle_y, label='Circle',color="hotpink")
```

# Triangle


```python
triangle_x = [5, 6, 7, 5]
triangle_y = [1, 3, 1, 1]

plt.plot(triangle_x,triangle_y,color='y', label='Triangle', marker='o')

plt.set_xlim(0, 8)
plt.set_ylim(0, 4)
plt.set_xlabel('X-axis')
plt.set_ylabel('Y-axis')

# Add legend
plt.legend()
plt.show()
```
