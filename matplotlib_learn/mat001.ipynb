# Figure
figure contains size of the window to plotting a drow.it can change default size 
of a plot 

```python
fig=plt.figure(figsize=(5,4))
#create a new axes for the figure
ax=fig.add_axes([2,2,2,2])
#adding the data to be plotted
ax.plot([2,3,4,5,6,6,7],
        [5,8,9,7,5,6,8])
plt.show()
```

```python
plt.plot([2,3,4,5,6,6,7],
        [5,8,9,7,5,6,8])
```

```python
# create multiple plot
fig=plt.figure(figsize=(5,4))
ax1=fig.add_axes([1,1,1,1])
ax2=fig.add_axes([1,0.42,0.5,0.5])
ax1.plot([12,17,20,18,62,45,88,10,15,18,11],
        [14,5,6,9,7,5,6,2,3,6,12])
ax2.plot([10,20,30,40,50,60],
         [1,2,3,4,5,6])
plt.show()
```

# subplot

```python
j=np.array([0,1,2,3])
k=np.array([3,8,1,10])
plt.subplot(1,2,1)
plt.plot(j,k)
plt.subplot(1,2,2)
plt.plot([1,2,3,4],j)

```

```python
x=np.array([0,1,2,4,10])
z=np.array([10,8,13,15,20])
plt.subplot(2,1,1)
plt.scatter(x,z)
plt.subplot(2,1,2)
plt.plot(z,x)
```

```python
A=np.array([[20,80,10,130,15],[6,8,10,15,18]])

B=np.array([[2,8,10,13,14],[6,8,10,15,18]])
C=np.array([[20,21,22,26,28],[1,3,5,5,8]])
D=np.array([[30,1,2,3,4],[5,7,11,15,20]])
E=np.array([[0,1,2,3,4],[11,15,16,18,20]])
F=np.array([[0,1,2,3,4],[5,12,18,20,22]])

#subploting
plt.subplot(2,3,1)
plt.plot(A)
plt.subplot(2,3,2)
plt.scatter(A,B, c="b",marker='*')
plt.subplot(2,3,3)
plt.plot(A,color="c")
plt.subplot(2,3,4)
plt.boxplot(E)
plt.subplot(2,3,5)
plt.plot(D)
plt.subplot(2,3,6)
plt.boxplot(F)
plt.show()
```
