```python
sns.lineplot(x="sepal_width",y="petal_width",data=df,hue="species")
#changing the theme to dark
sns.set_style(style="darkgrid")
sns.despine()
plt.show()
```

# remove the spines

```python
sns.lineplot(x="sepal_length",y="petal_length",data=df)
sns.despine()
plt.show()
```

# changing the figure size

```python
plt.figure(figsize=(3,4))

sns.lineplot(x="sepal_length",y="sepal_width",data=df)
sns.despine()
sns.set_context("paper")
plt.show()
```

```python
def plot():
    sns.lineplot(x="sepal_width",y="petal_length",data=df)
    
with sns.axes_style("darkgrid"):
    plt.subplot(211)
    plot()
plt.subplot(212)
plot()
```
