# Cutomizing the Line Plot

```python
#importing the libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#defining the dataset
x = np.arange(0,10,1)
y = 2*x + 5
#changing the figure
fig=plt.figure(figsize=(10,6))
#plotting the datapoints
plt.plot(x,y,linewidth = 2.0,linestyle = ":", color = 'r',alpha = 0.5,marker ='o')
plt.title("Line Plot Demo")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend(['line1'],loc='best')
plt.grid(True)
plt.show()
```

```python
a = np.arange(0,10,1)
b1 = 1*a + 10
b2 = 2*a + 10
b3 = 3*a + 10
a
#Change the figure
#fig = plt.figure(figsize=(10,6))
#plotting the data points

#Plot Graph 1
plt.subplot(1,3,1)
plt.plot(a,b1,linewidth = 2.0,linestyle=':', color='r',alpha=0.5,marker='o')
plt.title("Graph 1")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend(['line1'],loc='best')
plt.grid(True)

#Plot Graph 2
plt.subplot(1,3,2)
plt.plot(a,b2,linewidth = 2.0,linestyle=':', color='b',alpha=0.5,marker='o')
plt.title("Graph 2")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend(['line2'],loc='best')
plt.grid(True)

#Plot Graph 3
plt.subplot(1,3,3)
plt.plot(a,b3,linewidth = 2.0,linestyle=':', color='g',alpha=0.5,marker='o')
plt.title("Graph 3")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend(['line3'],loc='best')
plt.grid(True)


plt.show()
```

# How to create a Bar Plot

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#creating the dataset
data = {'apples':20,'Mangoes':15,'Lemon':30,'Oranges':10}
names = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (10,5))
#Creating the bar plot
plt.bar(names,values)
plt.show()
```

# Customizing the Bar Chart

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#creating the dataset
data = {'apples':20,'Mangoes':15,'Lemon':30,'Oranges':10}
names = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (10,5))
#Creating the bar plot
plt.barh(names[:2],values[:2],color="red")
plt.barh(names[2:],values[2:],color="orange")
plt.title("Bar Graph Demo")
plt.xlabel("Fruits")
plt.ylabel("Quantity")
plt.show()
```

# Customizing our Scatter Plot

```python
import matplotlib.pyplot as plt
%matplotlib inline

#Creating the dataset
a =[10,20,30,40,50,60,70,80]
b =[5,3,2,5,6,1,4,2]
x =[1,2,3,4,3,2,1,2]

#Creating the scatter Plot
plt.scatter(a,b, c='g', s=600,edgecolors='y',marker='o',alpha =0.5)
plt.scatter(a,x,c='b', s=400,edgecolors='r',marker='4',alpha =1)
plt.legend(['b','x'],loc='best')
plt.title("Scatter Plot Demo")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)
plt.savefig('Scatter Plot.png')
plt.show()
```

# How to create a Box Plot?

```python
import matplotlib.pyplot as plt
%matplotlib inline

#Data Preparation
total = [20,4,1,30,20,10,20,70,30,10]
order = [10,3,1,15,17,2,30,44,2,1]
discount = [30,10,20,5,10,20,50,60,20,45]
data = list([total, order,discount])
#Plotting the data
plt.boxplot(data, showmeans = True)
plt.title("Box Plot Demo")
plt.grid(True)
plt.show()
```

# How to create a violin plot?

```python
import matplotlib.pyplot as plt
%matplotlib inline

#Data Preparation
total = [20,4,1,30,20,10,20,70,30,10]
order = [10,3,1,15,17,2,30,44,2,1]
discount = [30,10,20,5,10,20,50,60,20,45]
data = list([total, order,discount])
#Plotting the data
plt.violinplot(data, showmeans = True, showmedians = True)
plt.title("Violin Plot Demo")
plt.grid(True)
plt.show()
```

# How to create a Pie Chart

```python
#importing the required libraries
import matplotlib.pyplot as plt
%matplotlib inline

#preparing the dataset
label = ['Dog','Cat','Wolf','Lion']
sizes = [50,45,60,80]

#creating the pie chart 
plt.pie(sizes, labels = label,autopct='%1.2f%%',shadow=True, startangle = 90,explode = (0.2,0,0.5,0))
plt.title("Pie Chart Demo")
plt.show()
```

# How to create a Doughnut Chart

```python
import matplotlib.pyplot as plt
%matplotlib inline

#define the dataset
group_names = ["GroupA","GroupB","GroupC"]
group_size = [20,30,50]
size_centre = [2,4,10,7,20]
a,b,c = [plt.cm.Blues,plt.cm.Reds,plt.cm.Greens]
#Creating the doughnut chart
pie1 = plt.pie(group_size,labels = group_names,autopct='%1i%%', radius = 1.5,colors = [a(0.5),b(1.5),c(1.0)])
pie2 = plt.pie(size_centre,radius =1,colors = 'r')
plt.show()
```

# How to create a Area Chart?

```python
import matplotlib.pyplot as plt
%matplotlib inline

#define the dataset
x=range(1,15)
y=[1,4,6,8,4,5,3,2,4,1,5,6,8,7]

#creating the area plot
plt.stackplot(x,y,colors = 'Black', alpha = 0.5)
#Merge area chart with the line chart
plt.plot(x,y,color = 'g')
plt.grid(True)
plt.show()
```
